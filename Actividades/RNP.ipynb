```python
# Actividad 8: Uso de software para búsqueda.
# Inteliencia Artificial y Redes Neuronales N4

# INTEGRANTES:
# ELIEZER RICARDO HERRERA OLIVA 1995385 IMTC
# CARLOS ANTONIO MACÍAS GONZÁLEZ 2127909 IMTC
# ALEK CAZAREZ MOYA 2006844 IMTC
# Axel Alonso Medrano Alanis 1581338

# Importar librerías necesarias
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from sklearn.metrics import accuracy_score

# Cargar el conjunto de datos MNIST
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()

# Normalizar los datos para que estén en el rango [0,1]
train_data, test_data = train_data / 255.0, test_data / 255.0

# Crear el modelo de red neuronal
model = Sequential([
    Flatten(input_shape=(28, 28)),    # Aplanar las imágenes de 28x28 píxeles
    Dense(128, activation='relu'),    # Capa oculta con 128 neuronas y función de activación ReLU
    Dense(64, activation='relu'),     # Segunda capa oculta con 64 neuronas
    Dense(10, activation='softmax')   # Capa de salida con 10 neuronas y función de activación softmax
])

# Compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entrenar el modelo
model.fit(train_data, train_labels, epochs=5, batch_size=32)

# Evaluar el modelo con el conjunto de datos de prueba
test_loss, test_accuracy = model.evaluate(test_data, test_labels)
print(f'Exactitud en el conjunto de prueba: {test_accuracy:.4f}')

# Realizar predicciones en el conjunto de datos de prueba
predicciones = model.predict(test_data[:5])   # Predice los primeros 5 datos de prueba ( EPOCAS USADAS  )

# Mostrar las predicciones y los valores reales
for i in range(5):
    print(f'Predicción: {predicciones[i].argmax()}, Valor Real: {test_labels[i]}')

# Importar librería para graficar
import matplotlib.pyplot as plt

# Realizar predicciones en el conjunto de datos de prueba
predicciones = model.predict(test_data[:5])   # Predice los primeros 5 datos de prueba

# Esta parte de aqui es solo un extra asi que si quieren quitenla pero ayuda a mejorar el codigo visualmente jaja

# Mostrar las predicciones, valores reales y la imagen de cada número
for i in range(5):
    plt.imshow(test_data[i], cmap='gray')  # Muestra la imagen en escala de grises
    plt.title(f'Predicción: {predicciones[i].argmax()}, Valor Real: {test_labels[i]}')
    plt.axis('off')  # Oculta los ejes para que se vea más limpio
    plt.show()
```
